```
#Settings
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('helloworld.xlsx')
df.rename(columns = {'Unnamed: 0' : 'date'}, inplace = True)
df = pd.pivot_table(df, index='date', aggfunc=np.sum)
```

```
def drawGraph(TitleList, data):
    print(TitleList)
    if len(TitleList)==0:
        print("No elements")
        return
    plt.figure(figsize = (50,10))
    graph = []
    graph.append(plt.subplot())
    graph[0].plot(df.index[:-7], df[TitleList[0]].values[:-7])
    print(TitleList[0])
    for i in range(1,len(TitleList)):
        graph.append(graph[0].twiny())
        graph[i].plot(df.index[:-7], df[TitleList[i]].values[:-7])
        print(TitleList[i])
```

```
drawGraph(list(df.columns), df)
```
